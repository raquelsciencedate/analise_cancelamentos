## Passo 4 
---
Agora finalmente iremos iniciar a análise de dados propriamente dita
vamos começar analisando a taxa de cancelamento dos dos planos de assiantura da empresa. Nesse momento iremos verificar qual a taxa de cancelamento e ver de onde ela vem. Código abaixo vamos ver quantos clientes cancelou e quantos não cancelou.

```python
display(tabela["cancelou"].value_counts())
```

Temos um número. A linha com valores **1.0** representa quem cancelou e **0.0** não cancelou. Mas para um leigo não fará sentido então iremos transformar em porcentagem
Formatando números: precisa do da função def a formatação em questão e :.2%, abaixo veremos que o lucro está abaixo do esperado.

```python
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

O .map("{:.2%}) atribui a lista uma operação que transforma os valores em **porcentagem**  descobrir o motivo de tantos cancelamentos

Parte 5 
Analise por partes. Duração, tipos ex: anual, trimestral e anual e duração de cada um. Analize logo após o porque  o mensal dura tão pouco já os anuais e trimestrais não são tão divergentes assim.

```python
display (tabela["duracao_contrato"].value_counts())
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.2%}".format))
```

Agrupando  e fazendo a média das informações da coluna 

```python
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
```

Aqui temos uma informação alarmante os contratos mensais tiveram 100% de cancelamento.
